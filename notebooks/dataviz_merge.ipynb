```python
#load gps coordinates
## testing with data_gps.
data_gps = pd.read_csv(external_data/'data_gps.csv')
data_gps['code_INSEE'] = ['0'+str(X) if len(str(X)) < 5 else X for X in data_gps['code_INSEE']]
data_gps.columns = ['CODE_INSEE','latitude','longitude']
```

```python
def plot_gps_column(data,column,xlim=(-6.5, 10.5), ylim=(41.5, 51.5)):
    #data = pd.merge(dataframe_total,data_gps,on='CODE_INSEE')
    fig, ax = plt.subplots(figsize = (10,8))
    index = data[column].isna()
    if index.sum():
        ax.scatter(data.longitude[index], data.latitude[index], c='k', s=20)
        sc = ax.scatter(data.longitude[-index], data.latitude[-index], c=data[column][-index], s=20,cmap='Spectral')
    else:
        sc = ax.scatter(data.longitude, data.latitude, c=data[column], s=20,cmap='Spectral')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_xlabel('longitude')
    ax.set_ylabel('latitude')
    ax.set_title("Column : {}, NaNs({})".format(column,index.sum()))
    fig.colorbar(sc)
    
```

```python
def fill_with_median(dataframe,column):
    # Get median by department
    df = dataframe.loc[:,['CODE_INSEE',column]]
    df['DEP'] = [str(X)[:2] for X in dataframe['CODE_INSEE']]
    TX_MEDIAN = df.groupby(['DEP']).median().reset_index()
    print("Example median",TX_MEDIAN.loc[0])
    
    # Fill with median
    df_total = pd.DataFrame(columns=['CODE_INSEE',column])
    for dep,median in zip(TX_MEDIAN.DEP,TX_MEDIAN[column]):
        df_department = df.loc[df.DEP == dep,['CODE_INSEE',column]].fillna(value=median)
        df_total = pd.concat([df_department,df_total])
        #print("Fill {} department".format(dep))
    return df_total
```

```python
fill = fill_with_median(tx_demande,'TX_RSA')
fill
```

```python
plot_gps_column( pd.merge(fill,data_gps,on="CODE_INSEE",how='left'),'TX_RSA')
```

```python
tx_demande = pd.read_csv(interim_data/'taux_couverture_mobile.csv')
fill_ = fill_with_median(tx_demande,'COUVERTURE_MOBILE')
plot_gps_column( pd.merge(fill_,data_gps,on="CODE_INSEE",how='left'),'COUVERTURE_MOBILE')
```
