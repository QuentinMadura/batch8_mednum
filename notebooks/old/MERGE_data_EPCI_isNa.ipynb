```python
import pandas as pd
from pathlib import Path
external_data = Path('../data/external/')
raw_data = Path('../data/raw/')
```

```python
df= pd.read_csv(external_data/ "MERGE_data.csv")
```

# Version booléenne du dataframe

```python
boolean_df = df.isna()
```

```python
boolean_df = boolean_df.drop(["CODE_INSEE", "LIBCOM", "DEP"], axis=1)
```

```python
# on renomme les colonnes pour simplifier la jointure
for col in boolean_df:
    boolean_df= boolean_df.rename(columns={col: str(col)+"_isNa"})
boolean_df
```

```python
# merge via l'index avec le df d'origine
df_flag = pd.merge(df, boolean_df, left_index=True, right_index=True)
```

```python
df_flag
```

## ajout des EPCI (code + libellé)

```python
# chargement des données sur les EPCI de l'INSEE
## https://www.insee.fr/fr/statistiques/fichier/2510634/Intercommunalite-Metropole_au_01-01-2020.zip
epci_xls = pd.ExcelFile(raw_data/ "Intercommunalité - Métropole au 01-01-2020.xlsx")
```

```python
df_epci = pd.read_excel(epci_xls, 'Composition_communale', skiprows=5)
```

```python
df_epci
```

```python
# renommage + selection de colonnes
df_epci = df_epci[["CODGEO", "EPCI", "LIBEPCI"]].rename(columns={"CODGEO": "CODE_INSEE"})
```

```python
df_epci
```

```python
# merge
df_demoday = pd.merge(df_epci, df_flag, on="CODE_INSEE")
```

```python
df_demoday
```

```python
df_demoday.to_csv("merge_data_epci_isNa.csv")
```
